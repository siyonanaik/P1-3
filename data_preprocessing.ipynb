```python
import pandas as pd
df = pd.read_csv('/nvidia.csv')
df
```

```python
# Drop unwanted columns
df = df.drop(['Dividends','Stock Splits'], axis=1)
df
```

```python
# Rename the Date column
df.rename(columns={'Unnamed: 0': 'Date'}, inplace=True)
df
```

```python
# Only date is needed, time can be removed
df['Date'] = df['Date'].str.slice(0, 10)
df
```

```python
# Check if there are any empty cells
df.isnull().any()
```

```python
# Check data types of each column
df.dtypes
```

```python
# Change data type of 'Date' column to datetime and 'Volume' column to int
df['Date'] = pd.to_datetime(df['Date'])
df['Volume'] = df['Volume'].astype(int)
```

```python
# Make sure the changes are made for the data type change
df.dtypes
```

```python
# Outlier check for 'Open', 'High', 'Low', 'Close' columns
import matplotlib.pyplot as plt
import numpy as np


all_data = [df['Open'], df['High'], df['Low'], df['Close']]

fig, ax = plt.subplots()
ax.boxplot(all_data)

ax.set_xticklabels(['Open', 'High', 'Low', 'Close'])
ax.set_title('Outlier Check')
ax.set_ylabel('Value')

plt.show()
```

*No outliers detected*

```python
# Outlier check for 'Volume' column
plt.boxplot(df['Volume'])

plt.xlabel("Volume")
plt.ylabel("Value")
plt.title("Outlier Check")

plt.show()
```

*Although there are outliers for the 'Volume' column, however this is acceptable as the stock market is volatile and they are in acceptable range. It does not look like a wrong entry, but just normal stock market values.*
